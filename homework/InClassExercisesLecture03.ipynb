2. qq plot

```python
from scipy.stats import norm
from scipy.stats import uniform
from scipy.stats import cauchy
import statsmodels.api as sm

random_norm = norm.rvs(size=1000)
random_uniform = uniform.rvs(size=1000)
random_cauchy = cauchy.rvs(size=1000)
```

```python
sm.qqplot(random_norm)
```

```python
sm.qqplot(random_uniform)
```

```python
sm.qqplot(random_cauchy)
```
